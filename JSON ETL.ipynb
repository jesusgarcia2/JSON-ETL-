```python
#Import libraries
import sys
sys._enablelegacywindowsfsencoding()
import json
import os

def main():
        #Define in and out routes
        dirin = r"C:\Users\Jesús García\Desktop\Python ETL\news_in"
        dirout = r"C:\Users\Jesús García\Desktop\Python ETL\news_out"
         
        print("Start JSON ETL (Python)")
        print("--------------------------")
        count = 1
        for file in os.listdir(dirin):
            if file.endswith(".json"):
                with open(os.path.join(dirin, file),'r', encoding='utf8') as f:
                    data = json.load(f)
                    data_new = {}
                    data_new['id'] = str(count)
                    data_new['published'] = data['published']
                    data_new['text'] = data['text']
                    data_new['title'] = data['title']
                    data_new['author'] = data['author']
                with open(os.path.join(dirout, file), 'w', encoding='utf8') as f:
                    json.dump(data_new, f)
                    print("Document number %d extracted successfully" % count)
                    count = count+1
                    
       
 
# This is the standard boilerplate that calls the main() function
if __name__ == '__main__':
    main()
    print("--------------------------")
    print("End JSON ETL (Python)")
```
